```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
df = pd.read_csv("D:\DEV VELLY\Datasets\creditcard.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.fillna('ffill', inplace = True)
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace = True)
```

```python
print('No Frauds', round(df['Class'].value_counts()[0]/len(df) *100, 2), '% of the dataset')
print('Frauds', round(df['Class'].value_counts(0)[1]/len(df) * 100, 2), '% of the dataset')
```

```python
f, ax = plt.subplots(1, 2, figsize = (18, 6))
df['Class'].value_counts().plot.pie(autopct = '%1.1f', ax = ax[0], shadow = True)
ax[0].set_title('distribution')
ax[0].set_ylabel('')
sns.countplot(x = 'Class', data = df, ax = ax[1])
ax[1].set_title('Class')
plt.show()
```

```python
df
```

```python
from sklearn.preprocessing import RobustScaler, StandardScaler
# scaler = RobustScaler()
scaler = StandardScaler()
```

```python
df['Amount'] = scaler.fit_transform(df['Amount'].values.reshape(-1,1))
df['Time'] = scaler.fit_transform(df['Time'].values.reshape(-1, 1))
df.head()
```

```python
X = df.drop("Class", axis = 1)
y = df["Class"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 2 )
```

```python
y.value_counts()
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print("Accuracy : ", accuracy_score(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test,y_pred)
print(cm)

plt.figure(figsize = (6, 5))
sns.heatmap(cm, annot = True, fmt = 'd', cmap = 'Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_auc_score, roc_curve

y_pred_proba = model.predict_proba(X_test)[:,1]
auc = roc_auc_score(y_test, y_pred_proba)
print('AUC: %.3f' % auc)

fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)

plt.figure(figsize = (6, 5))
plt.plot(fpr, tpr, label = f" AUC = {auc:.3f}")
plt.plot([0, 1], [0, 1], '--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.grid(True, alpha = 0.3)
plt.show()
```

```python
from imblearn.over_sampling import RandomOverSampler
oversample = RandomOverSampler(sampling_strategy = 'minority')
X_randomover, y_randomover = oversample.fit_resample(X_train, y_train)
```

```python
y_randomover.value_counts()
```

```python
model.fit(X_randomover, y_randomover)
y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy : %.3f%%" % (accuracy))
```

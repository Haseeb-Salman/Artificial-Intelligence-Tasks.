```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")
```

1. Read Student file

```python
df = pd.read_csv("student_data_clean.csv")
df.head()
```

2. Data information

```python
df.shape
df.info()
df.describe()
```

3. Check Missing Values

```python
df.isnull().sum()
```

4. Checking columns

```python
df.columns
```

5. Fill Null values

```python
df[["CGPA","Attendance_Rate"]] = df[["CGPA","Attendance_Rate"]].fillna(
    df[["CGPA","Attendance_Rate"]].mean())
df.isnull().sum()
```

6. Fill String Null Values


```python
df["Fees_Paid"] = df["Fees_Paid"].fillna(df["Fees_Paid"].mode()[0])
df["Fees_Paid"]
```

7. Average CGPA of Students

```python
df.isnull().sum()
```

```python
df["CGPA"].mean()
```

8. Students with CGPA greater than 3.5

```python
df[df["CGPA"] > 3.5]
```

9. Average study hours per day

```python
df["Study_Hours_Per_Day"].mean()
```

10. Students by depaertments

```python
df["Department"].value_counts()
```

11. Average Marks By Gender

```python
df.groupby("Gender")['Final_Marks'].mean()
```

12. Correlation Heatmap

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

13. Scatter Plot

```python
plt.figure(figsize = (8, 6))
sns.scatterplot(x = "Study_Hours_Per_Day", y = "CGPA", data = df)
plt.title("Study Hour VS CGPA")
plt.show()
```

14. Line Plot

```python
plt.figure(figsize = (8, 6))
sns.lineplot(x = "Semester", y = "CGPA", data = df)
plt.title("Semester VS CGPA")
plt.show()
```

15. Bar Plot

```python
plt.figure(figsize = (8, 6))
sns.barplot(x = "Department", y = "CGPA", data = df)
plt.title("Average CGPA by Department")
plt.show()
```

16. Count plot

```python
plt.figure(figsize = (8, 6))
sns.countplot(x = "Gender", data = df)
plt.title("Number Of Students By GFender")
plt.show()
```

17. BOX plot

```python
plt.figure(figsize = (8, 6))
sns.boxplot(x = "Department", y = "CGPA", data = df)
plt.title("CGPA Distribution by Gender")
plt.show()
```

18. Violin Plot


```python
plt.figure(figsize = (8, 6))
sns.violinplot(x = "Gender", y = "CGPA", data = df)
plt.title("CGPA Distribution By Gender")
plt.show()
```

19. Histogram plot

```python
plt.figure(figsize = (8, 6))
sns.histplot(df["CGPA"], bins = 20, kde = True)
plt.title("CGPA Distribution")
plt.show()
```

20. KDE plot


```python
plt.figure(figsize = (8, 6))
sns. kdeplot(df["Attendance_Rate"], fill = True)
plt.title("Attendance Rate Density")
plt.show()
```

21. Pair Plot

```python
sns.pairplot(df)
plt.show()
```

22. Strip Plot

```python
plt.figure(figsize = (8, 6))
sns.stripplot(x = "Gender", y = "CGPA", data = df)
plt.title("CGPA VS Gender")
plt.show()
```

23. Swarm Plot

```python
plt.figure(figsize = (8, 6))
sns.swarmplot(x = "Gender", y = "CGPA", data = df)
plt.title("CGPA VS Gender")
plt.show()
```

24. Regression Plot

```python
plt.figure(figsize = (8, 6))
sns.regplot(x = "Study_Hours_Per_Day", y = "CGPA", data = df)
plt.title("Regression : Study Hours VS CGPA")
plt.show()
```

# Input Data

```python
name = input("What is your name? ")  # Ask the user for their name and store it in the variable 'name'
birth_year = int(input("What year were you born? "))  # Ask the user for their birth year and store it in the variable 'birth_year'
age = int(input("What is your current age? "))  # Ask the user for their current age and store it in the variable 'age' 
year = 100 - age + birth_year  # Calculate the year when the user will turn 100 years old
lucky_number = float(input("Enter your lucky number: "))  # Ask the user for their lucky number and store it in the variable 'lucky_number'
```

# Example Output

```python
print("What is your name? ", name)  #Shows the name of the user
print("What year were you born? ", birth_year)  #Shows the birth year of the user
print("What is your current age? ", age)  #shows the user age
print("Enter you lucky number: ", lucky_number)
```

# Personal Summary

```python
print('-'* 4, 'Personal Assistant Summary', '-'* 4 )
print("Hello,", name, '!')# Greet the user by their name
print("You were born in ", birth_year, ".")  # Print the user's birth year
print("You are currently ", age, " years old.")  # Print the user's current age
print("You will turn 100 in the year ", year, ".")  # Print the year when the user will turn 100 years old
print("Your lucky number is ", lucky_number, ".")  # Print the user's lucky number
print('-'*36)
```

# Program 1: Variables (Name, Age, City)

```python
name = "Haseeb"
age = 20
city = "Gujranwala"

print("Name:", name)
print("Age:", age)
print("City:", city)
```

# Program 2: Data Types

```python
num_int = 10
num_float = 5.5
text = "Hello"
flag = True

print(type(num_int))    # int
print(type(num_float))  # float
print(type(text))       # str
print(type(flag))       # bool
```

# Program 3: Arithmetic Operations

```python
a = 10
b = 5

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
```

# Program 4: Comparison Operators

```python
x = 8
y = 12

print("x > y:", x > y)
print("x < y:", x < y)
print("x == y:", x == y)
```

# Program 5: Type Conversion

```python
int_val = 7
float_val = 9.8

print("Integer to Float:", float(int_val))
print("Float to Integer:", int(float_val))
```

# Program 6: Student Information

```python
student_name = "Ali"
roll_no = 101
marks = 88

print("Name:", student_name)
print("Roll No:", roll_no)
print("Marks:", marks)
```

# Program 7: Assignment Operators

```python
num = 10
num += 5
print("After += :", num)

num -= 3
print("After -= :", num)

num *= 2
print("After *= :", num)
```

# Program 8: Boolean Operations

```python
p = True
q = False

print("p and q:", p and q)
print("p or q:", p or q)
print("not p:", not p)
```

The Number Cruncher 

```python
for i in range(1, 20):      #using for loop to iterate numbers from 1 to 20
    if i%3 == 0:        # if condition to check thew number is completely divideable on 3
        print("Fizz")   
    elif i%5 == 0:
        print("Buzz")  # if condition to check thew number is completely divideable on 5
    elif i%3 == 0 and i%5 == 0:  # if condition to check thew number is completely divideable on both
        print("FizzBuzz")
    else:
        print(i)
```

# CountDown

```python
num = int (input("Enter an Number : "))  #input an number From the user
while num >= 1:  # while loop for iteration from input number to 1
    if num > 10:  #if statement to checknumber is greater than 10 or not
        break   # to terminate an loop
    else:   
        print("Countdown: ",num)
        num = num - 1      # decrement 1 from num
if num <=10:
    print("Blast Off!")
```

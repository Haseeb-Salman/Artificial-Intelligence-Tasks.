```python
# Objective: Perform basic exploratory data analysis using Pandas

# 1. Import libraries
import pandas as pd
import numpy as np
```

```python
# 2. Create dictionary
data = {
    'Name': ['Alice','Bob','Charlie','Diana','Eve','Frank'],
    'Age': [25,30,np.nan,35,28,np.nan],
    'Score': [85,92,78,np.nan,88,95],
    'Grade': ['A','A','B','C','B','A']
}
```

```python
# 3. Create DataFrame
df = pd.DataFrame(data)
```

```python
# 4. Basic EDA
print("First 3 rows:")
print(df.head(3))

print("\nShape of DataFrame:")
print(df.shape)

print("\nColumns:")
print(df.columns)

print("\nData Types:")
print(df.dtypes)
```

```python
# 5. Missing Data Handling
print("\nMissing Values:")
print(df.isnull())

# Fill Age with mean
df['Age'].fillna(df['Age'].mean(), inplace=True)

# Drop rows with missing Score
df = df.dropna(subset=['Score'])

print("\nData after handling missing values:")
print(df)
```

```python
# 6. Selection
print("\nSelect Name and Score:")
print(df[['Name', 'Score']])

print("\nRows where Score > 85:")
print(df[df['Score'] > 85])

# Set Name as index
df.set_index('Name', inplace=True)
print("\nDataFrame with Name as index:")
print(df)
```

```python
# 7. Operations
print("\nUnique Grades:")
print(df['Grade'].unique())

print("\nGrade Counts:")
print(df['Grade'].value_counts())

print("\nSort by Score Descending:")
print(df.sort_values(by='Score', ascending=False))
```

```python
# Objective: Demonstrate broadcasting and universal functions

import numpy as np

# 1. Create arrays
arr1 = np.array([[1, 2, 3], [4, 5, 6]])
arr2 = np.array([10, 20, 30])
```

```python
# 2. Broadcasting Operations
addition = arr1 + arr2
subtraction = arr1 - arr2
multiplication = arr1 * arr2
division = arr1 / arr2

print("Addition:\n", addition)
print("Subtraction:\n", subtraction)
print("Multiplication:\n", multiplication)
print("Division:\n", division)
```

```python
# Broadcasting Explanation:
# Broadcasting allows NumPy to perform operations on arrays of different shapes.
# arr2 is automatically expanded to match the shape of arr1 (2x3),
# so operations can be performed element-wise.
```

```python
# 3. Universal Functions
arr3 = np.array([1, 4, 9, 16, 25])

print("\nSquare Root:", np.sqrt(arr3))
print("Exponential:", np.exp(arr3))
print("Sine:", np.sin(arr3))
```

```python
# 4. Array operations on arr4
arr4 = np.array([[5, 12, 3], [8, 2, 7]])

print("\nSum of all elements:", np.sum(arr4))
print("Max in each column:", np.max(arr4, axis=0))
print("Min in each row:", np.min(arr4, axis=1))
```

```python
# 5. Type casting arr1 to float
arr1_float = arr1.astype(float)
print("\narr1 converted to float:\n", arr1_float)
```

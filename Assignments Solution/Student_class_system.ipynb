The Student Class System

```python
# base class Person
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    # Display information

    def display_info(self):
        print("\n--- Person ---")
        print("Name : ", self.name)
        print("AGE : ", self.age)
```

```python
# Derived class Student
class Student(Person):
    def __init__(self,name, age, student_id, major):
        super().__init__(name, age)
        self.student_id = student_id
        self.major = major
    
    # Display information

    def display_info(self):
        print("\n--- Student ---")
        print("Name : ", self.name)
        print("Age : ", self.age)
        print("Student ID : ", self.student_id)
        print("Major : ", self.major)
```

```python
# Derived class GraduateStudent
class GraduateStudent(Student):
    def __init__(self, name, age, student_id, major, thesis_topic):
        super().__init__(name, age, student_id, major)
        self.thesis_topic = thesis_topic
    
    # Display information

    def display_info(self):
        print("\n--- Graduate Student ---")
        print("Name:", self.name)
        print("Age:", self.age)
        print("Student ID:", self.student_id)
        print("Major:", self.major)
        print("Thesis Topic:", self.thesis_topic)
```

```python
# Main Program: Create objects
p = Person("John Doe", 45)
s = Student("Jane Smith", 20, "S12345", "Computer Science")
g = GraduateStudent("Bob Johnson", 27, "G67890", "Physics", "Quantum Computing")
```

```python
# Call methods
p.display_info()
s.display_info()
g.display_info()
```

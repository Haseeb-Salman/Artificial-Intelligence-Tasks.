```python
# ----------- Functions -----------
def arithmetic_operations(a, b):
    print("\nArithmetic Operations:")
    print("Addition:", a + b)
    print("Subtraction:", a - b)
    print("Multiplication:", a * b)
    print("Division:", a / b)


def check_number(num):
    print("\nConditional Statement:")
    if num > 0:
        print("Number is Positive")
    elif num < 0:
        print("Number is Negative")
    else:
        print("Number is Zero")


def print_list(items):
    print("\nFor Loop Example (List Items):")
    for item in items:
        print(item)


def countdown(n):
    print("\nWhile Loop Example (Countdown):")
    while n > 0:
        print(n)
        n -= 1
```

```python
# ----------- Main Program -----------

# Data Types
name = "Haseeb"
age = 21
height = 5.9
is_student = True

numbers_list = [10, 20, 30, 40, 50]
numbers_tuple = (1, 2, 3, 4, 5)
student_dict = {"name": name, "age": age, "city": "Gujranwala"}

print("----- Data Types -----")
print(type(name))
print(type(age))
print(type(height))
print(type(is_student))
print(type(numbers_list))
print(type(numbers_tuple))
print(type(student_dict))
```

```python
# Operators
a = 15
b = 5

arithmetic_operations(a, b)

print("\nComparison Operators:")
print("a > b:", a > b)
print("a < b:", a < b)
print("a == b:", a == b)
print("a != b:", a != b)

print("\nLogical Operators:")
x = True
y = False
print("x and y:", x and y)
print("x or y:", x or y)
print("not x:", not x)

print("\nMembership Operators:")
print("20 in list:", 20 in numbers_list)
print("100 not in list:", 100 not in numbers_list)
```

```python
# Conditional Statement
check_number(-5)
```

```python
# Loops
print_list(numbers_list)
countdown(5)
```

```python
# Dictionary Loop
print("\nDictionary Data:")
for key, value in student_dict.items():
    print(key, ":", value)


print("\nProgram Finished Successfully!")
```

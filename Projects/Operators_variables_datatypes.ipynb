```python
# 1. Variables and Data Types
name = "Haseeb"
age = 21
height = 5.8
is_student = True

# List, Tuple, Dictionary
numbers_list = [10, 20, 30, 40, 50]
numbers_tuple = (1, 2, 3, 4, 5)
student_dict = {
    "name": name,
    "age": age,
    "city": "Gujranwala"
}

# Display Data Types
print("----- Data Types -----")
print("Name:", type(name))
print("Age:", type(age))
print("Height:", type(height))
print("Is Student:", type(is_student))
print("List:", type(numbers_list))
print("Tuple:", type(numbers_tuple))
print("Dictionary:", type(student_dict))
```

```python
a = 15
b = 4

print("\n----- Arithmetic Operations -----")
print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Modulus:", a % b)
print("Power:", a ** b)
```

```python
# 3. Comparison Operators
print("\n----- Comparison Operators -----")
print("a > b:", a > b)
print("a < b:", a < b)
print("a == b:", a == b)
print("a != b:", a != b)
```

```python
# 4. Logical Operators
x = True
y = False

print("\n----- Logical Operators -----")
print("x and y:", x and y)
print("x or y:", x or y)
print("not x:", not x)
```

```python
# 5. Assignment Operators
num = 10
print("\n----- Assignment Operators -----")
print("Initial value:", num)

num += 5
print("After += :", num)

num *= 2
print("After *= :", num)

num -= 4
print("After -= :", num)
```

```python
# 6. Membership Operators
print("\n----- Membership Operators -----")
print("20 in numbers_list:", 20 in numbers_list)
print("100 not in numbers_list:", 100 not in numbers_list)
```

```python
# 7. Final Summary Output
print("\n----- Student Information -----")
for key, value in student_dict.items():
    print(key, ":", value)

print("\nProgram Finished Successfully!")
```

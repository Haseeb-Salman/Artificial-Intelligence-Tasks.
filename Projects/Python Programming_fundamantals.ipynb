```python
# -------- Functions --------
def arithmetic_operations(a, b):
    print("\nArithmetic Operations")
    print("Addition:", a + b)
    print("Subtraction:", a - b)
    print("Multiplication:", a * b)
    print("Division:", a / b)


def check_even_odd(num):
    if num % 2 == 0:
        print("Number is Even")
    else:
        print("Number is Odd")


def print_list(items):
    print("\nList Items:")
    for item in items:
        print(item)

def countdown(n):
    print("\nCountdown:")
    while n > 0:
        print(n)
        n -= 1
```

```python
# -------- Main Program --------
print("----- Python Basics Project -----")

# Input / Output
name = input("Enter your name: ")
age = int(input("Enter your age: "))
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
```

```python
# Data Types
height = 5.9
is_student = True

numbers_list = [10, 20, 30, 40, 50]
numbers_tuple = (1, 2, 3, 4, 5)
student_dict = {"name": name, "age": age}
numbers_set = {1, 2, 3, 4, 5}

print("\nData Types:")
print(type(name))
print(type(age))
print(type(height))
print(type(is_student))
print(type(numbers_list))
print(type(numbers_tuple))
print(type(student_dict))
print(type(numbers_set))
```

```python
# Operators
arithmetic_operations(num1, num2)

print("\nComparison Operators")
print(num1 > num2)
print(num1 < num2)
print(num1 == num2)

print("\nLogical Operators")
print((num1 > 0 and num2 > 0))
print((num1 > 0 or num2 > 0))
```

```python
# Conditional Statement
print("\nEven/Odd Check")
check_even_odd(num1)
```

```python
# Loops
print_list(numbers_list)
countdown(5)
```

```python
# Dictionary Loop
print("\nDictionary Data:")
for key, value in student_dict.items():
    print(key, ":", value)

print("\nProgram Finished Successfully!")
```

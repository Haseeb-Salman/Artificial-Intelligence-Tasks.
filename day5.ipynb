```python
class Vehicle:
    def move(self):
        print("Vehicle moves")
class Car(Vehicle):
    def move(self):
        print("Car drives")
class Bike(Vehicle):
    def move(self):
        print("Bike rides")
class Bus(Vehicle):
    def move(self):
        print("Bus transports")
Vehicle = [Car(), Bike(), Bus()]
for v in Vehicle:
    v.move()
```

```python
class Shapes:
    def area(self):
        print("Calculating area")

class Circle(Shapes):
    def __init__(self, radius):
        self.radius = radius
    def area(self):
        print(f"Area of Circle: {3.14 * self.radius **2}")

class Rectangle(Shapes):
    def __init__(self, width, length):
        self.width = width
        self.length = length    
    def area(self):
        print(f"Area of Rectangle : {self.length * self.width }")

class Triangle(Shapes):
    def __init__(self, base, height):
        self.base = base
        self.height = height
    def area(self):
        print(f"Area of Triangle : {0.5 * self.base * self.height}")

sh = Shapes()
sh.area()
circle = Circle(5)
circle.area()
rectangle = Rectangle(5, 6)
rectangle.area()
triangle = Triangle(4, 9)
triangle.area()
```

```python
class Employee:
    def calculate_salary(self):
        print("Calculating salary")
class FullTimeEmployee(Employee):
    def __init__(self, fixed_salary):
        self.fixed_salary = fixed_salary
    def calculate_salary(self):
        return self.fixed_salary
class PartTimeEmployee(Employee):
    def __init__(self, hourly_rate, hours_worked):
        self.hourly_rate = hourly_rate
        self.hours_worked = hours_worked
    def calculate_salary(self):
        return self.hourly_rate * self.hours_worked
class Intern(Employee):
    def __init__(self, stipend):
        self.stipend = stipend
    def calculate_salary(self):
        return self.stipend
emp = Employee()
while True:
    emp.calculate_salary()
    print("Enter Fixed Salary.....")
    fsalary = int(input("Enter fixed salary for full-time employee: "))
    if fsalary <= 0:
        print("Invalid salary")
        break
    else: 
        full_time_emp = FullTimeEmployee(fsalary)
        print(f"Full-time employee salary: {full_time_emp.calculate_salary()}")
    print("Enter Hourly Rate.....")
    hrate = int(input("Enter hourly rate for part-time employee: "))
    print("Enter Hours Worked.....")
    hworked = int(input("Enter hours worked for part-time employee: "))
    if hrate <= 0 or hworked <= 0:
        print("Invalid hourly rate or hours worked")    
        break
    else: 
        part_time_emp = PartTimeEmployee(hrate, hworked)
        print(f"Part-time employee salary: {part_time_emp.calculate_salary()}")
    print("Enter Stipend.....")
    stipend = int(input("Enter stipend for intern: "))
    if stipend <= 0:
        print("Invalid stipend")
        break
    else:
        intern = Intern(stipend)
        print(f"Intern salary: {intern.calculate_salary()}")
```

```python
class Customer:
    def get_discount(self, amount):
        pass
class RegularCustomer(Customer):
    def get_discount(self, amount):
        discount = amount * 0.05
        return amount-discount
class PremiumCustomer(Customer):
    def get_discount(self, amount):
        discount = amount * 0.10
        return amount-discount  
class VIPCustomer(Customer):
    def get_discount(self, amount):
        discount = amount * 0.15
        return amount-discount

Customer = [RegularCustomer(), PremiumCustomer(), VIPCustomer()]
amount = 1000
for customer in Customer:
    print("Final Payable Amount for " + str(customer.__class__.__name__) + " is $" + str(customer.get_discount(amount)))
```

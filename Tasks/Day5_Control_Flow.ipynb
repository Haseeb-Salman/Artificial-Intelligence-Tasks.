## If, elif, else Statements
Used to execute different blocks of code based on conditions.
Helps programs make decisions dynamically.

```python
# 10 Examples of if-elif-else
x = 10
if x > 5:
    print('Greater than 5')

y = -3
if y > 0:
    print('Positive')
elif y < 0:
    print('Negative')
else:
    print('Zero')

age = 18
if age >= 18:
    print('Adult')

num = 0
if num == 0:
    print('Zero')

score = 85
if score >= 90:
    print('A')
elif score >= 80:
    print('B')

temp = 30
if temp > 25:
    print('Hot')

a, b = 5, 10
if a < b:
    print('a is smaller')

is_raining = True
if is_raining:
    print('Take umbrella')

name = ''
if not name:
    print('Name empty')

n = 7
if n % 2 == 0:
    print('Even')
else:
    print('Odd')
```

## Break, Continue, Pass
Control loop execution: break exits, continue skips, pass does nothing.
Useful for fine‑grained control inside loops.

```python
# 10 Examples
for i in range(10):
    if i == 5:
        break
    print(i)

for i in range(5):
    if i == 2:
        continue
    print(i)

for i in range(3):
    pass

while True:
    break

for ch in 'hello':
    if ch == 'l':
        continue
    print(ch)

for i in range(5):
    if i == 3:
        pass
    print(i)

for i in range(10):
    if i > 7:
        break

for i in range(5):
    if i % 2 == 0:
        continue
    print(i)

if True:
    pass

for i in []:
    break
```

## Nested Conditionals
Conditionals placed inside other conditionals.
Used when decisions depend on multiple levels of checks.

```python
# 10 Examples
x = 10
if x > 0:
    if x % 2 == 0:
        print('Positive Even')

age = 20
if age >= 18:
    if age >= 21:
        print('Can drink in US')

n = 5
if n > 0:
    if n < 10:
        print('Single digit')

a, b = 5, 3
if a > b:
    if a - b > 1:
        print('Much bigger')

score = 92
if score >= 90:
    if score == 100:
        print('Perfect')

temp = 15
if temp > 0:
    if temp < 20:
        print('Cool')

if True:
    if True:
        print('Nested True')

x = -5
if x < 0:
    if x % 2 != 0:
        print('Negative Odd')

num = 8
if num % 2 == 0:
    if num % 4 == 0:
        print('Divisible by 4')

value = 50
if value > 10:
    if value < 100:
        print('Between 10 and 100')
```

## Conditional (Ternary) Expression
A one‑line shortcut for simple if‑else assignments.
Improves code conciseness for small decisions.

```python
# 10 Examples
x = 10
print('Even' if x % 2 == 0 else 'Odd')

a, b = 5, 7
max_val = a if a > b else b

age = 16
status = 'Adult' if age >= 18 else 'Minor'

n = -1
print('Positive' if n > 0 else 'Non‑positive')

flag = True
msg = 'Yes' if flag else 'No'

score = 40
result = 'Pass' if score >= 50 else 'Fail'

name = 'Ali'
print(name if name else 'Unknown')

temp = 30
weather = 'Hot' if temp > 25 else 'Cool'

x = 0
print('Zero' if x == 0 else 'Non‑zero')

num = 9
parity = 'Even' if num % 2 == 0 else 'Odd'
```

## Nested Loops
Loops inside other loops.
Commonly used for grids, tables, and combinations.

```python
# 10 Examples
for i in range(2):
    for j in range(2):
        print(i, j)

for a in 'AB':
    for b in '12':
        print(a, b)

for i in range(3):
    for j in range(i):
        print('*', end='')
    print()

matrix = [[1,2],[3,4]]
for row in matrix:
    for val in row:
        print(val)

for i in range(2):
    for j in range(2):
        for k in range(1):
            print(i,j,k)

for i in range(2):
    for j in range(2):
        pass

for x in [1,2]:
    for y in [3,4]:
        print(x+y)

for i in range(1,3):
    for j in range(1,3):
        print(i*j)

for i in range(2):
    for j in range(2):
        if i==j:
            print('Equal')

for i in range(2):
    for j in range(2):
        print('#', end='')
    print()
```

## List Comprehension
Concise way to create lists using loops and conditions.
Often faster and more readable than traditional loops.

```python
# 10 Examples
squares = [x*x for x in range(5)]
evens = [x for x in range(10) if x%2==0]
chars = [c for c in 'AI']
pairs = [(x,y) for x in [1,2] for y in [3,4]]
neg = [-x for x in [1,2,3]]
lengths = [len(w) for w in ['hi','hello']]
bools = [x>2 for x in range(5)]
cubes = [x**3 for x in range(4)]
filtered = [x for x in range(10) if x>5]
double = [2*x for x in range(5)]
```

## Iterators & Iterables
Iterables are objects you can loop over; iterators produce items one at a time.
Used for memory‑efficient data processing.

```python
# 10 Examples
lst = [1,2,3]
it = iter(lst)
print(next(it))

s = 'AI'
it2 = iter(s)
print(next(it2))

for v in lst:
    print(v)

t = (1,2)
print(list(iter(t)))

r = range(3)
print(list(r))

d = {'a':1,'b':2}
for k in d:
    print(k)

gen = (x*x for x in range(3))
print(next(gen))

print(sum(iter([1,2,3])))

print(any(iter([0,1,0])))

print(all(iter([1,2,3])))
```

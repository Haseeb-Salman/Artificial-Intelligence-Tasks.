## While Loop — 10 Examples

```python
i = 1
while i <= 5:
    print(i)
    i += 1
```

Prints numbers from 1 to 5 using a loop.
The loop runs until the condition becomes False.

```python
count = 5
while count > 0:
    print(count)
    count -= 1
```

Calculates the sum of first five natural numbers.
Adds each number during iteration.

```python
num = 2
while num <= 10:
    print(num)
    num += 2
```

Prints even numbers from 2 to 10.
Increments by 2 each time.

```python
x = 1
while x < 4:
    print("Hello")
    x += 1
```

Prints a word multiple times.
Stops after 3 repetitions.

```python
i = 1
while i <= 5:
    if i == 3:
        break
    print(i)
    i += 1
```

Stops the loop when i equals 3.
Demonstrates use of break.

```python
i = 0
while i < 5:
    i += 1
    if i == 3:
        break
    print(i)
```

Skips printing number 3.
Continue moves to next iteration.

```python
password = ""
while password != "1234":
    password = "1234"
print("Access Granted")
```

Simulates password checking.
Loop runs until correct input.

```python
i = 1
while i <= 3:
    j = 1
    while j <= 2:
        print(i, j)
        j += 1
    i += 1
```

Example of nested while loops.
Prints pairs of numbers.

```python
i = 1
while True:
    print(i)
    if i == 3:
        break
    i += 1
```

Infinite loop controlled by break.
Stops manually when condition met.

## Functions — 10 Examples

```python
def greet():
    print("Hello") 
greet()
```

Defines a simple function with no parameters.
Calls the function to execute it.

```python
def add(a, b):
    return a + b
print(add(2, 3))
```

Takes two inputs and returns their sum.
Demonstrates return values.

```python
def square(x):
    print(x*x)
square(4)
```

Prints square of a number.
Uses a single parameter.

```python
def info(name, age):
    print(name, age)
info("Ali", 20)
```

Function with multiple parameters.
Displays provided arguments.

```python
def greet(name="User"):
    print("Hello", name)
greet()
```

Shows default parameter value.
Uses default when no argument given.

```python
def even(n):
    return n % 2 == 0
print(even(6))
```

Returns True if number is even.
Uses boolean return.

```python
def factorial(n):
    f = 1
    for i in range(1, n+1):
        f *= i
    return f
print(factorial(5))
```

Computes factorial using a loop.
Returns the final result.

```python
def show(*nums):
    print(nums)
show(1,2,3)
```

Accepts variable number of arguments.
Stored as a tuple.

```python
def person(**data):
    print(data)
person(name="Ali", age=20)
```

Accepts keyword arguments.
Stored as a dictionary.

```python
def outer():
    def inner():
        print("Inner function")
    inner()
outer()
```

Example of nested functions.
Inner function called inside outer.

## Variable Scope — 10 Examples

```python
x = 10
def show():
    print(x)
show()
```

Accesses global variable inside function.
Global variables are readable inside functions.

```python
def test():
    y = 5
    print(y)
test()
```

Local variable defined inside function.
Cannot be accessed outside.

```python
x = 10
def change():
    x = 5
    print(x)
change()
print(x)
```

Local variable shadows global variable.
Global value remains unchanged.

```python
x = 10
def modify():
    global x
    x = 20
modify()
print(x)
```

Uses global keyword to modify global variable.
Changes original value.

```python
def outer():
    z = 7
    def inner():
        print(z)
    inner()
outer()
```

Inner function accesses outer variable.
Demonstrates enclosing scope.

```python
def counter():
    count = 0
    count += 1
    print(count)
counter()
```

Local variable resets each call.
Does not retain value.

```python
a = 1
b = 2
def swap():
    global a, b
    a, b = b, a
swap()
print(a, b)
```

Modifies multiple global variables.
Shows global declaration.

```python
def f():
    x = 1
    def g():
        nonlocal x
        x = 2
    g()
    print(x)
f()
```

Uses nonlocal to modify enclosing variable.
Applies to nested functions.

```python
name = "Ali"
def display():
    print(name)
display()
```

Reads global string inside function.
No modification occurs.

```python
def demo():
    p = 3
    print(p)
demo()
```

Simple example of local scope.
Variable exists only during function execution.

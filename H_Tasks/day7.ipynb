```python
d = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age' : [20, 30, 35, 40],
    'Salary': [20000, 30000, 40000, 50000],
    'Department':['HR', 'IT', 'IT', 'Finance']
})

print(d.shape,'\n')
print(d.info())
print('\n', d.describe())
```

```python
print(d['Name'])
print('\n',d[['Name','Age']])
print('\n')
print(d)
print('\n')
print("First row by position")
print(d.iloc[0])
print('\n')
print("First Row by index")
print(d.loc[0])
print('\n')
print("Slicing Rows")
print(d.iloc[1:3])
print('\n')
print("Conditional Selection")
print(d[d['Age']>28])
print('\n')
print(d[(d['Age']>25) & (d['Department']== 'IT')])
```

```python
df = pd.DataFrame({
    'Name' : ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank'],
    'Age' : [25, 30, 28, 26, 39, 45],
    'Salary' : [10000, 20000, 30000, 40000, 50000, 60000],
    'Department' : ['HR', 'IT', 'IT', 'HR', 'Finance', 'IT'],
    'Experience' : [2, 5, 8, 3, 2, 6]
})

print('Head')
print(df.head(2))
print('\n')
print('tail')
print(df.tail(2))
print('\n')
print('Unique')
print(df['Department'].unique())
print('\n')
print('nunique')
print(df['Department'].nunique())
```

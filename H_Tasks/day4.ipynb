```python
class bank():
    def __init__(self):
       
        self.balance = 0
        pass

    def set_pin(self, pin):
        self.pin = pin


    def deposit(self, amount):
        self.balance += amount
        print(f"{amount} deposited. New balance: {self.balance}")

    def withdraw(self, amount):
        if amount > self.balance or self.balance == 0 or amount <= 0 :
            print("INPUT VALID AMOUNT TO WITHDRAW.")
        else:
            self.balance -= amount
            print(f"{amount} withdrawn. New balance: {self.balance}")

    def check_balance(self):
        print(f"Current balance: {self.balance}")
    
    def menu(self):
        print("SET YOUR PIN FOR LOGIN.....")
        pin = input("Enter your PIN: ")
        print("PIN set successfully. Please login to access your account.")

        print("ENTER YOUR PIN TO LOGIN : ")
        Pin = input()

        if Pin == pin:
         while True:
            print("\nMenu:")
            print("1. Check Balance")
            print("2. Deposit")
            print("3. Withdraw")
            print("4. Exit")
            choice = input("Enter your choice: ")

            if choice == '1':
                print("Please enter your PIN to check balance.")
                pin1 = input("Enter PIN : ")
               
                if pin1 == pin:
                    print("Verifying your PIN......")
                    self.check_balance()
                else:
                    print("Incorrect PIN. Access denied.")
            elif choice == '2':
                print("Please enter your PIN to Deposit Amount.")
                pin1 = input("Enter PIN : ")

                if pin1 == pin:
                    print("Verifying your PIN......")
                    print("Enter Amount To Deposit: ")
                    amount = float(input("Enter amount to deposit: "))
                    self.deposit(amount)
                else:
                    print("Incorrect PIN. Access denied.")
            
            elif choice == '3':
                print("Please enter your PIN to Withdraw Amount.")
                pin1 = input("Enter PIN : ")

                if pin1 == pin:
                    print("Verifying your PIN......")
                    amount = float(input("Enter amount to withdraw: "))
                    self.withdraw(amount)
                else:
                    print("Incorrect PIN. Access denied.")
            elif choice == '4':
                print("Exiting...")
                break
            else:
                print("Invalid choice. Please try again.")
        else:
            print("Invalid Pin!......")


bank1 = bank()
bank1.menu()
```

```python
class Playlist():
    def __init__(self, name):
        self.songs=[]
    def add_song(self, songs):
        self.songs.append(songs)
        print(f"{songs} added to the playlist.")
    def remove_song(self,songs):
        if songs in self.songs:
            self.songs.remove(songs)
            print(f"{songs} removed from the playlist.")
    def show_songs(self):
        print("Songs in the Playlist:")
        for song in self.songs:
            print(f" - {song}")
playlist = Playlist("Favourite Songs-")
playlist.add_song("Song 1")
playlist.add_song("Song 2")
playlist.show_songs()
del playlist.songs[1]
playlist.remove_song("Song 1")
playlist.show_songs()

    
```
